### 스텝1 - 자료와 패키지 불러오기, 데이터 구성하기

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
import numpy as np
from tqdm import tqdm
from sklearn.model_selection import KFold, StratifiedKFold
```

```python
path = '/home/piai/Downloads/big_data/'

df_raw = pd.read_csv(path + '유방암.csv')
df_raw.head()
```

```python
df_raw.info()
```

```python
df_raw.isnull().sum()
```

```python
# 데이터에 대해서 학습, 검증, 테스트 데이터로 분할한다.
X = df_raw.drop('diagnosis', axis = 1)
Y = df_raw['diagnosis']

trainX, testX, trainY, testY = train_test_split(X, Y, test_size = 0.3, random_state = 1234)
trainX, validX, trainY, validY = train_test_split(trainX, trainY, test_size = 0.4, random_state = 1234)
```

```python
# 분리된 데이터셋의 (행, 열)을 살펴보면 다음과 같다.
print(trainX.shape)
print(trainY.shape)
print(validX.shape)
print(validY.shape)
print(testX.shape)
print(testY.shape)
```

### 스텝2 - SVC모델 생성하기

```python
svm_uncustomized = SVC(random_state = 1234)
svm_uncustomized.fit(trainX, trainY)

print('학습 데이터셋 정확도: {:.3f}'.format(svm_uncustomized.score(trainX, trainY)))
print('테스트 데이터셋 정확도: {:.3f}'.format(svm_uncustomized.score(testX, testY)))
```

```python
# 공교롭게도 학습과 테스트 데이터셋에 대한 정확도가 0.908로 같게 나왔다.
# 하이퍼 파라미터 튜닝 과정을 거쳐서 정확도를 개선시켜보도록 한다.
```

```python
# C 파라미터 튜닝
train_accuracy_c, valid_accuracy_c = [], []

C = [c for c in np.arange(0.01,10,0.05)]


for param_c in tqdm(C):
    svm = SVC(C = param_c, random_state = 1234)
    svm.fit(trainX, trainY)
    train_accuracy_c.append(svm.score(trainX,trainY))
    valid_accuracy_c.append(svm.score(testX, testY))

df_accuracy_c = pd.DataFrame()
df_accuracy_c['C'] = C
df_accuracy_c['TrainAccuracy'] = train_accuracy_c
df_accuracy_c['ValidAccuracy'] = valid_accuracy_c
```

```python
plt.plot(C, train_accuracy_c, linestyle = '-', label = 'TrainAccuracy')
plt.plot(C, valid_accuracy_c, linestyle = '--', label = 'ValidAccuracy')
plt.ylabel('accuracy'); plt.xlabel('C')
plt.legend()
```

```python
df_accuracy_c[80:110]
```

```python
# 위의 그래프와 데이터프레임으로 부터 C값이 4와 5 사이일 경우 훈련 데이터와 트레이닝 데이터의 정확도가 우수하다.
# 데이터프레임을 면밀히 살펴보았을 때, 4.16이상인 값 부터 일정 구간동안 같은 정확도를 가진다.
# 따라서, C를 4.16으로 설정한다.
```

```python
# Gamma 파라미터 튜닝
train_accuracy_gamma, valid_accuracy_gamma = [], []

gammas = [gamma for gamma in np.arange(0.01,10,0.05)]


for gamma in tqdm(gammas):
    svm = SVC(C = 4.16, gamma = gamma, random_state = 1234)
    svm.fit(trainX, trainY)
    train_accuracy_gamma.append(svm.score(trainX,trainY))
    valid_accuracy_gamma.append(svm.score(testX, testY))

df_accuracy_gamma = pd.DataFrame()
df_accuracy_gamma['C'] = gammas
df_accuracy_gamma['TrainAccuracy'] = train_accuracy_gamma
df_accuracy_gamma['ValidAccuracy'] = valid_accuracy_gamma
```

```python
df_accuracy_gamma.round(4)
```

```python
plt.plot(gammas, train_accuracy_gamma, linestyle='-', label = 'Train Accuracy')
plt.plot(gammas, valid_accuracy_gamma, linestyle='--', label = 'Valid Accuracy')
plt.ylabel('accuracy'); plt.xlabel('gamma')
plt.legend()
```

```python
# gamma의 변화는 정확도에 영향을 미치지 않는다.
# 따라서, gamma 파라미터는 디폴트로 세팅하도록 한다.
```

### 스케일 변환

```python
v_feature_name = df_raw.columns

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_raw)
df_scaled = pd.DataFrame(df_scaled, columns = v_feature_name)

df_scaled.head()
```

```python
# 데이터 분할 후, 다시 모델을 구성해보도록 한다.
# 데이터에 대해서 학습, 검증, 테스트 데이터로 분할한다.

X_scaled = df_scaled.drop('diagnosis', axis = 1)
Y_scaled = df_scaled['diagnosis']

# Y_scaled는 0 혹은 1의 값을 가지는 바이너리 형태여야 하므로 수치를 변경해준다.
# 악성을 1로, 양성을 0으로 재코딩해준다.
Y_scaled = Y_scaled.map({Y_scaled.value_counts().index[0]:1,Y_scaled.value_counts().index[1]:0})
trainX, testX, trainY, testY = train_test_split(X_scaled, Y_scaled, test_size = 0.3, random_state = 1234)
trainX, validX, trainY, validY = train_test_split(trainX, trainY, test_size = 0.4, random_state = 1234)
```

```python
# C 파라미터 튜닝
train_accuracy_c, valid_accuracy_c = [], []

C = [c for c in np.arange(0.01,10,0.05)]

for param_c in tqdm(C):
    svm = SVC(C = param_c, random_state = 1234)
    svm.fit(trainX, trainY)
    train_accuracy_c.append(svm.score(trainX,trainY))
    valid_accuracy_c.append(svm.score(testX, testY))

df_accuracy_c = pd.DataFrame()
df_accuracy_c['C'] = C
df_accuracy_c['TrainAccuracy'] = train_accuracy_c
df_accuracy_c['ValidAccuracy'] = valid_accuracy_c
```

```python
df_accuracy_c[20:50]
```

```python
plt.plot(C, train_accuracy_c, linestyle = '-', label = 'TrainAccuracy')
plt.plot(C, valid_accuracy_c, linestyle = '--', label = 'ValidAccuracy')
plt.ylabel('accuracy'); plt.xlabel('C')
plt.legend()
```

```python
# 위의 시각화 자료와 데이터프레임을 통해 c=2의 값을 설정해준다.
# 해당 값에서 트레이닝 데이터와 검증 데이터에 대한 정확도가 좋기 때문이다.
# 트레이닝 정확도는 1 이후로 일정값에 수렴하고, 검증데이터셋은 2 이후 다소 낮아지는 경향이 있다.
```

```python
# Gamma 파라미터 튜닝
train_accuracy_gamma, valid_accuracy_gamma = [], []

gammas = [gamma for gamma in np.arange(0.01,10,0.05)]


for gamma in tqdm(gammas):
    svm = SVC(C = 2, gamma = gamma, random_state = 1234)
    svm.fit(trainX, trainY)
    train_accuracy_gamma.append(svm.score(trainX,trainY))
    valid_accuracy_gamma.append(svm.score(testX, testY))

df_accuracy_gamma = pd.DataFrame()
df_accuracy_gamma['C'] = gammas
df_accuracy_gamma['TrainAccuracy'] = train_accuracy_gamma
df_accuracy_gamma['ValidAccuracy'] = valid_accuracy_gamma
```

```python
df_accuracy_gamma.round(4)
```

```python
plt.plot(gammas, train_accuracy_gamma, linestyle='-', label = 'Train Accuracy')
plt.plot(gammas, valid_accuracy_gamma, linestyle='--', label = 'Valid Accuracy')
plt.ylabel('accuracy'); plt.xlabel('gamma')
plt.legend()
```

```python
# gamma값이 높아질 수록 오히려 검증 데이터셋의 정확도가 낮아지는 경향을 보인다.
# 반면, 
# 따라서, Gamma의 값을 0.01으로 설정하도록 한다.
```

### 스텝3 - 최종 모델 선택

```python
svc_final = SVC(gamma=0.01, C=2,random_state = 1234)
svc_final.fit(trainX,trainY)

y_pred = svc_final.predict(testX)

print('훈련 데이터 정확도: {:.3f}'.format(svc_final.score(trainX,trainY)))
print('테스트 데이터 정확도: {:.3f}'.format(svc_final.score(testX,testY)))
print('혼동 행렬: \n{}'.format(confusion_matrix(testY,y_pred)))
```

```python
# 혼동행렬로부터 구한 분류 예측 결과는 다음과 같다.
print('Accuracy(정확도) is: {:.2f}'.format((56+104)/(56+10+1+104)))
print('Precision(정밀도) is: {:.2f}'.format((104)/(10+104)))
print('Recall(재현율) is: {:.2f}'.format((104)/(1+104)))
```

# 번외) GridSearch로 최적의 파라미터 조합 찾기

```python
# 주어진 데이터셋은 0과 1을 구분하는 분류와 관련된 데이터셋이다. 
# 따라서, 클래스의 불균형인 상태로 데이터가 split되는 상황에 대처하기 위해 CV진행 시 Stratified CV를 사용하는 것이 적합하다.
kfold = StratifiedKFold(n_splits=5)
```

```python
Cs = [c for c in np.arange(0.01,10,0.05)]
GAMMAs = [gamma for gamma in np.arange(0.01,10,0.05)]

estimator = SVC(random_state = 1234)

param_grid = {'C': Cs, 'gamma':GAMMAs}

grid_SVC = GridSearchCV(estimator, param_grid, scoring = 'accuracy', n_jobs = -1, cv = kfold)
grid_SVC.fit(trainX,trainY)
```

```python
print('best estimator model:\n{}'.format(grid_SVC.best_estimator_))
print('best parameter:\n{}'.format(grid_SVC.best_params_))
print('best score:\n{}'.format(grid_SVC.best_score_.round(3)))
```

```python
# 그리드 서치 결과, 최적의 파라미터 조합은 C= 2.71, gamma = 0.01로 산출되었다.
# 정확도는 0.987로, 앞서 혼동행렬로부터 구한 0.94보다 개선된 수치이다.
```
