### 1번 과제 - 28페이지

다음 데이터는 자동차의 모델과 연비(mpg)에 대한 자료를 조사한 Table입니다. 이 범주형 데이터에 대해 counts, percents, cumulative counts, cumulative percents를 계산 바랍니다.(data.set : mycars.csv)

```python
import pandas as pd
import numpy as np
from scipy import stats

df_cars = pd.read_csv('/home/piai/Downloads/data_stats/mycars.csv')
```

```python
df_cars.head()
```

```python
df = df_cars[['model','mpg']]

# counts
count = df['model'].value_counts().sort_index()


# cumulative counts
cumcnt = np.cumsum(count)


# percents
percent = count/sum(count) * 100


# cumulative percents
cumpct = np.cumsum(percent)
```

```python
df_result = pd.DataFrame({'Count':count,'CumCnt':cumcnt,'Percent':percent,'Cumpct':cumpct}).sort_index()
df_result.columns.name = 'model'
df_result
```

```python
# counts
count2 = df['mpg'].value_counts().sort_index()

# cumulative counts
cumcnt2 = np.cumsum(count2)


# percents
percent2 = count2/sum(count2) * 100


# cumulative percents
cumpct2 = np.cumsum(percent2)
```

```python
df_result2 = pd.DataFrame({'Count':count2,'CumCnt':cumcnt2,'Percent':percent2,'Cumpct':cumpct2}).sort_index()
df_result2.columns.name = 'mpg'
df_result2
```

### 2번 과제 - 45페이지

어떤 자동차 1리터당 주행거리가 평균 12km, 표준편차 3km인 정규분포를 이룬다.  
1) 1리터를 가지고 9km 이상 15km 이하를 달릴 확률은 얼마인가요?

```python
mean_ = 12
dev_ = 3

# 1)
# prob1: 누적확률 15이하 값 계산

x1 = 15
prob1 = stats.norm.cdf(x1, mean_, dev_)

# prob2: 누적확률 9이하 값 계산

x2 = 9
prob2 = stats.norm.cdf(x2, mean_, dev_)

print('1리터를 가지고 9km 이상 15km 이하를 달릴 확률은 얼마인가요?\n답은', round(prob1 - prob2,3), '입니다.')
```

2) 1리터를 가지고 15km 이상을 달릴 확률은 얼마인가요?

```python
# prob3: 누적확률 15이상 값 계산

x3 = 15
prob3 = 1 - stats.norm.cdf(x3, mean_, dev_)

print('1리터를 가지고 15km 이상을 달릴 확률은 얼마인가요?\n답은', round(prob3,3),'입니다.')
```
