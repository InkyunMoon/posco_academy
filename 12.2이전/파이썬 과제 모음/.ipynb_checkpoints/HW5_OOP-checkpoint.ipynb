```python
class Set:
# 1 PAGE
    def __init__(self,lst=[]):
        ls = []
        for i in lst:
            if i not in ls:
                ls.append(i)
        self.lst = ls
    
    def __repr__(self): # 내장함수 set처럼 Set([1,2])만 해주어도 {1,2}반환하는 기능
        res = '{'+', '.join(list(map(str,self.lst)))+'}'
        return res
        
    def add(self, elem):
        if elem not in self.lst:
            self.lst.append(elem)
        return self
    
    def discard(self, elem):
        if elem in self.lst:
            self.lst.remove(elem)
        return self.lst
    
    def clear(self):
        self.lst = []
        return self.lst
    
    def __len__(self):
        count = 0
        for i in self.lst:
            count += 1
        return count
# 2 PAGE

    def __str__(self):
        temp = str(self.lst).replace('[','{')
        temp = temp.replace(temp[temp.rfind(']')],'}')
        return temp
    
    def __contains__(self,elem):
        if elem in self.lst:
            return True
        else:
            return False
        
    def __le__(self,other):
        for i in self.lst:
            num = other.lst.count(i)
            if num == 0:
                return False
        return True
    
    def __ge__(self,other):
        for i in other.lst:
            num = self.lst.count(i)
            if num == 0:
                return False
        return True
# 3 PAGE
    def __or__(self, other):
        return Set(self.lst + other.lst)
    
    def __and__(self, other):
        temp = []
        for i in self.lst:
            if i in other.lst:
                temp.append(i)
        return Set(temp)
    
    def __sub__(self, other):
        temp = []
        for i in self.lst:
            if i not in other.lst:
                temp.append(i)
        return Set(temp)
    
    def __ior__(self, other):
        for i in other.lst:
            if i not in self.lst:
                self.lst.append(i)
        return self

    def __iand__(self, other):
        temp = []
        for i in self.lst:
            if i in other.lst:
                temp.append(i)
        self.lst = temp
        return self
    
    def __isub__(self, other):
        temp = []
        for i in self.lst:
            if i not in other.lst:
                temp.append(i)
        self.lst = temp
        return self
```

```python
a = Set([1,2,3,4])
b = Set([1,2,3,4])

print(a)
print(b)
print()

a.discard(4)
b.discard(1)
print(a)
print(b)
print()

print(len(a))
print(1 in a)
print(1 in b)
print()

print(a | b)
print(a & b)
print(a - b)
print()

print(a <= b)
print(a <= a | b)
print(a >= b)
print(a >= a & b)
print()

b.clear()
print(b)
print()

a = Set([1,2,3])
b = Set([3,4])
address_a = id(a)
a |= b
print(a)
print(address_a == id(a))
print()

a = Set([1,2,3])
b = Set([3,4])
address_a = id(a)
a &= b
print(a)
print(address_a == id(a))
print()

a = Set([1,2,3])
b = Set([3,4])
address_a = id(a)
a -= b
print(a)
print(address_a == id(a))
print()
```
